Name	Description
- 1	Year: 2004-2008
- 2	Month: 1-12
- 3	DayofMonth: 1-31
- 4	DayOfWeek: 1 (Monday) - 7 (Sunday)
- 5	DepTime: Actual departure time (local, hhmm)
- 6	CRSDepTime: Scheduled departure time (local, hhmm)

- 9	UniqueCarrier: Unique carrier code

- 11	TailNum: Plane tail number
- 12	ActualElapsedTime: In minutes
- 13	CRSElapsedTime: In minutes
- 14	AirTime: In minutes
- 15	ArrDelay: Arrival delay, in minutes
- 16	DepDelay: Departure delay, in minutes
- 17	Origin: Origin IATA airport code
- 18	Dest: Destination IATA airport code
- 19	Distance: In miles
- 20	TaxiIn: Taxi in time, in minutes
- 21	TaxiOut: Taxi out time in minutes
- 22	Cancelled: Was the flight cancelled?
- 23	CancellationCode: Reason for cancellation (A = carrier, B = weather, C = NAS, D = security)
- 24	Diverted: 1 = yes, 0 = no
- 25	CarrierDelay: In minutes
- 26	WeatherDelay: In minutes
- 27	NASDelay: In minutes
- 28	SecurityDelay: In minutes
- 29	LateAircraftDelay: In minutes

> Only 9368 flights are recorded as cancelled with it averaging 1.9% of all flights.  It would be interesting to get a stat on the average distance a flight goes that is cancelled vs okay.

```python
sb.relplot(data = df, x = 'Distance', y='Cancelled', kind="line")
```

```python
plt.hist2d(data = df, x='Cancelled', y="Distance")
plt.colorbar()
```

>Very Curious to see that the shorter flights were more likely to be cancelled.  Going to dig deeper into the Cancelled flights to see whats going on and who the biggest culprit is.

```python
#Moving data with any cancellations in it into a new dataframe.
df_c = df[df['Cancelled']> 0]
print(df_c.info())
```

```python
df_c = df_c.drop(['ArrDelay', 'DepTime', 'ArrTime', 'DepDelay', 'ActualElapsedTime', 'AirTime'], axis=1)
```

```python
df_c.head()
```

```python
#Replace Letter symbols with actually words for better analysis
df_c.replace({'CancellationCode': {'A':'Carrier', 'B':'Weather', 'C':'NAS','D':'Security'}}, inplace=True)#A = carrier, B = weather, C = NAS, D = security
df_c['CancellationCode'].value_counts()
```

```python
#Saving Cancelled data frame to csv for quick future analysis
df_c.to_csv('Cancelled-Record.csv', sep='\t', encoding='utf-8', index=False)
```
